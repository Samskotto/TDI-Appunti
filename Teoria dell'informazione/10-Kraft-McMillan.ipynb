Abbiamo trovato che **Huffman** √® ottimale tra tutti i codici istantanei. Ma nel caso in cui volessimo trovare un codice ottimo ma **non istantaneo**? <mark>Non esiste un codice univocamente decodificabile che lavori meglio di **Huffman**</mark>.
___
Per dimostrare il teorema seguente abbiamo bisogno di definire l'estensione $k$-esima di un codice come
$$
    C_k : \mathbb{X}^k \rightarrow D^+
$$
Questa funzione ci permette di codificare un blocco di dimensione $k$, ad esempio:
$$
    \text{Andrea}\rightarrow C_6(\text{Andrea}) = C(A)\cdot C(n) \cdot C(d) \dots
$$
Inoltre definiamo:
 - $l_{C_k} = \sum_{i=1}^{k} l_c(x_i)$
 - $l_{max} = \max_{i=1,\dots,k} l_c(x_i)$, quindi la codifica del singolo simbolo pi√π grande.
## Teorema
Data la sorgente $\mathbb{X} = \{x_1,\dots,x_m\}$, e dati $m$ interi $l_1,\dots,l_m > 0$ che mi rappresentano le lunghezze dei simboli del codice, esiste un codice **univocamente decodificabile** con base $D > 1$ per la sorgente 
$$
\begin{align}
    \mathbb{c}:\mathbb{X}\rightarrow D^+ \quad \text{tale che}\quad l_c(x_i) = l_i \quad \forall i = 1,\dots,m
\end{align}
$$
**se e solo se vale**
$$
\begin{align}
    \sum_{i=1}^{m} D^{-l_i} \leq 1
\end{align}
$$
### Dimostrazione
1. Se vale $\sum_{i=1}^{m} D^{-l_i} \leq 1$ vuol dire che la disugualgianza di Kraft √® rispettata e che quindi le lunghezze espresse sono sia di un **codice istantaneo che √® un tipo particolare di codice univocamente decodificabile**.
2. Dimostriamo l‚Äôaltro lato dell'implicazione, ovvero **che dato un codice univocamente decodificabile** vale la disuguaglianza, partiamo definendo:
    $$
       \Big(\sum_{x \in \mathbb{X}} D^{-l_c(x)}\Big)^k
    $$
   Possiamo riscrivere la potenza di una sommatoria in questo modo, vediamolo per $k = 2$:
   $$
   \begin{align}
       \Big(\sum_{i} a_i\Big)^{2} &= \sum_{i} a_i \sum_{j} a_j\\
       &=\sum_{i}\sum_{j} a_i\cdot a_j
   \end{align}
   $$
   Possiamo quindi riscrivere $\Big(\sum_{x \in \mathbb{X}} D^{-l_c(x)}\Big)^k$ come:
   $$
   \begin{align}
       \Big(\sum_{x \in \mathbb{X}} D^{-l_c(x)}\Big)^k &= \sum_{x_1 \in \mathbb{X}}\dots\sum_{x_k \in \mathbb{X}} D^{-l_c(x_1)}\cdot\dots\cdot D^{-l_c(x_k)} \quad &(1)\\
       &= \sum_{x_1 \in \mathbb{X}}\dots\sum_{x_k \in \mathbb{X}} D^{-(l_c(x_1)+l_c(x_2)+\dots+l_c(x_k))} \quad &(2)\\
       &= \sum_{x_1,\dots,x_k \in X^{k} } D^{-(l_{C_k}(x_1,\dots,x_k))} \quad &(3)\\
   \end{align}
   $$
   Abbiamo riscritto la sommatoria passando alla codifica dell'estensione che prende un blocco di $k$-oggetti, di conseguenza utilizziamo l'insieme $\mathbb{X}^k$ che contiene tutte le possibili disposizioni di $k$-simboli della sorgente.

Notiamo due fatti importanti: 
- la lunghezza minima di codifica di una parola di $k$-simboli √® pari a $k$, questo perch√® se abbiamo una parola di lunghezza $k$ e assegniamo ad ogni simbolo $1$-bit di codifica (che √® la **codifica minima**) otteniamo $k$ come lunghezza minima. Non possiamo scendere al di sotto perch√® dovremmo introdurre la parola vuota $\alpha$ che non appartiene a nessun codice.
- la lunghezza massima di codifica di una parola di lunghezza $k$ √® data da $k\cdot l_{max}$ (*Semanticamente* utilizzo per **ogni simbolo che compone la mia parola la lunghezza massima**).

Ora raggruppiamo tutte le parole di $k$-simboli di lunghezza totale $n$, otteniamo quindi delle partizioni dell'insieme $\mathbb{X}^k$
$$
\begin{align}
    {\mathbb{X}_{n}^{k}} = \{(x_1,\dots,x_k) \in \mathbb{X}^{k}\quad \text{tali che}\quad l_{C_k}(x_1,\dots,x_k) = n\} 
\end{align}
$$
Riscriviamo la sommatoria come la sommatoria di tutte le partizioni possibili che ci sono sull'insieme ${\mathbb{X}_{n}^{k}}$, quindi aggiungiamo una sommatoria che ci permette di "muoverci" tra **le varie lunghezze possibili delle parole**.
$$
\begin{align}
\sum_{n=1}^{k\cdot l_{\max}}\sum_{(x_1,\dots,x_k) \in {\mathbb{X}_{n}^{k}}} D^{-(l_{C_k}(x_1,\dots,x_k))}\quad &(1)\\
\sum_{n=1}^{k\cdot l_{\max}}\sum_{(x_1,\dots,x_k) \in {\mathbb{X}_{n}^{k}}} D^{-(l_{C_k}(x_1,\dots,x_k))}\quad &(2)\\
\sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot\sum_{(x_1,\dots,x_k) \in {\mathbb{X}_{n}^{k}}} 1 \quad &(3)\\
\end{align}
$$
Abbiamo estratto visto che $l_{C_k}(x_1,\dots,x_k) = n$ questo perch√® stiamo lavorando sulla partizione (basta guardare la definizione), inoltre dato che $D^{-n}$ √® indipendente dalla sommatoria possiamo portarlo a sinistra lasciando questa sommatoria $\sum_{(x_1,\dots,x_k) \in {\mathbb{X}_{n}^{k}}} 1$, ma questa sommatoria sostanzialmente sta calcolando il numero di elementi presenti nella partizione quindi $|{\mathbb{X}_{n}^{k}}|$. Riscriviamo
$$
\begin{align}
   \sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot |{\mathbb{X}_{n}^{k}}|
\end{align}
$$
Ora applichiamo una maggiorazione a questa quantit√†: dato che sappiamo che $\mathbb{c}$ √® univocamente decodificabile per definizione anche la sua estensione $C_k$ √® **non singolare quindi UD**. Il codominio di questa funzione deve essere pi√π grande o al massimo uguale al dominio:
$$
    C_k : {\mathbb{X}_{n}^{k}} \rightarrow D^+
$$
Ma dato che $|{\mathbb{X}_{n}^{k}}| \leq D^{n}$:
$$
\begin{align}
   &\sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot |{\mathbb{X}_{n}^{k}}| \leq  \sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot D^{n} \quad &(1)\\
   &\sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot |{\mathbb{X}_{n}^{k}}| \leq  \sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot D^{n} \quad &(2)\\
    &\sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot |{\mathbb{X}_{n}^{k}}| \leq  \sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot D^{n} \quad &(3)\\
&\sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot |{\mathbb{X}_{n}^{k}}| \leq  \sum_{n=1}^{k\cdot l_{\max}} 1 \quad &(4)\\
&\sum_{n=1}^{k\cdot l_{\max}} D^{-n}\cdot |{\mathbb{X}_{n}^{k}}| \leq  k\cdot l_{\max} \quad &(5)\\
\end{align}
$$
Abbiamo quindi torvato questa disuguaglianza, se chiamamo $M = \sum_{x \in \mathbb{X}} D^{-l_c(x)}$ e tracciamo il grafico di $M^k$:
- se $M > 1$ ho un esponenziale crescente, che da un certo punto $k_0$ rende falsa la relazione;
- se $0 \leq M \leq 1$ ho un esponenziale decrescente, che da un certo punto $ùëò_0$ rende vera la
relazione.
Quindi da un certo $k_0$ in poi la funzione $M^k$ sta sotto a $k\cdot l_{\max}$, in questo modo abbiamo dimostrato che vale la disuguaglianza di **Kraft**.
___
*Semanticamente*: **i limiti ottenuti sulle lunghezze ottimali delle parole di codice** restano validi anche quando si estende la classe dei codici ammessi a quella di tutti i codici univocamente decodificabili.

Dato che dobbiamo trovare un codice univocamente decodificabile con base $D$ possiamo utilizzare la disuguaglianza di **Kraft/McMillan**. 
$$
\begin{align}
&\sum_{i=1}^{6} D^{-l_i} \leq 1 \\
=& 2\cdot \Big(D^{-1}+ D^{-2}+D^{-3}\Big) \leq 1 
\end{align}
$$
Per $D = 3$ ottengo
$$
    2 \cdot \Big(\frac{1}{3} + \frac{1}{9} + \frac{1}{27}\Big) = 0.96 \leq 1
$$
