## Notazioni preliminari 
1. **Chain rule**
   $$
   \begin{align}
        &H(X,Y) = H(X) + H(Y|X) = H(Y) + H(X|Y) \quad &(1) \\
        \\
        &\text{su spazio condizionato}\\
        \\
        &H(X,Y|Z ) = H(X|Z) + H(Y|X,Z) = H(Y|Z) + H(X|Y,Z) \quad &(2) 
   \end{align}
   $$
2. **Diagrammi di Venn**
   
## Disuguaglianza di Fano
*Semanticamente*: Vogliamo spedire un carattere $X$ al ricevente $Y$ sul canale. Se il canale presenta **rumore** il carattere che abbiamo spedito potr arrivare al ricevente con degli errori, di conseguenza il ricevente avr un messaggio diverso dall'originale. Vogliamo quindi utilizzare una funzione $g(Y)$ che risalga al carattere originale $X$, e vogliamo calcolare qual  **la probabilit di errore** che questa funzione commette.
<div style="background-color: #D1D7FF; padding: 10px; border-radius: 5px;text-align: center; font-size:16px;">
<b>Definizione:</b> Siano $X$ e $Y$ due variabili aleatorie che prendono valori in $\mathbb{X},\mathbb{Y}$ insiemi finiti. Sia $g: \mathbb{Y} \rightarrow \mathbb{X}$ una funzione che mappa i valori di $\mathbb{Y}$ nei valori di $\mathbb{X}$. 
    
Sia <b>$P_e$</b> la proabilit di erroe nel <b>predire</b> $X$ quando uso $g$:
$$
    P_e = P(g(Y) \neq X)
$$
Allora 
$$
    P_e \geq \frac{H(X|Y) - 1}{\log |\mathbb{X}|}
$$
</div>

### Dimostrazione 
Introduciamo una variabile aleatoria *Bernoulliana* $E$ tale che
$$
E = \begin{cases}
    &1 \quad \text{se} \quad g(Y) \neq X\\
    &0 \quad \text{se} \quad g(Y) = X\\
\end{cases}
$$
Per la **Chain rule** sugli spazi condizionati sappiamo che possiamo scrivere:
$$
    H(E,X|Y)  = H(E|Y) + H(X|E,Y) \quad \text{oppure} \quad H(E,X|Y) = H(X|Y) + H(E|X,Y)
$$
Possiamo fare le seguenti affermazioni:
1. $H(E|X,Y) = 0$, questo perch conoscendo $X,Y$ posso calcolare $g(Y)$ e vedere se  uguale a $X$, e quindi sapere subito il valore di $E$ (quindi non ho bisogno di informazioni aggiuntive per descrivere $E$).
2. $H(E|Y) \leq H(E)$, questo lo si nota facendo il **diagramma di Venn**.

Consideriamo ora $H(X|E,Y)$ e lo riscriviamo come:
$$
\begin{align}
    H(X|E,Y) &= P(E = 0) \cdot H(X|E=0,Y) + P(E= 1) \cdot H(X| E=1,Y) \quad &(1)\\
    \\
    &= (1 - P_e) \cdot H(X|E=0,Y) + P_e \cdot H(X| E=1,Y) \quad &(2)\\
\end{align}
$$
Dal momento che $E$  una *Bernoulliana* sappiamo che $P(E = 0) = 1 - P_e$ mentre $P(E = 1) = P_e$ dove $P_e$  la probabilit di errore della funzione $g$. A questo punto per notiamo che $(1 - P_e) \cdot H(X|E=0,Y) = 0$ perch se sappiamo che la variabile $E = 0$ vuol dire che non ci sono stati errori e che quindi $g(Y) = X$, di conseguenza l'entropia  pari a zero perch non abbiamo informazioni aggiuntive per descrivere questo evento.
$$
\begin{align}
   H(X|E,Y) &= 0 + P_e \cdot H(X| E=1,Y) \quad &(3)\\
\end{align}
$$
Ora scriviamo
$$
    P_e \cdot H(X| E=1,Y) + H(E) \geq H(X|Y) + 0 
$$
Sapendo che **l'entropia** ha come **upper bound** il $\log_Dm$ (dove $m$  la cardinalit di $X$) quindi possiamo applicare una maggiorazione a $H(X| E=1,Y)\leq \log m - 1$, sottraiamo $-1$ perch per $E = 1$ sto considerando tutti i valori di $X$ tranne quello corretto. Mentre $H(E)$ per una variabile aletoria *Bernoulliana*  sempre $H(E) \leq 1$. Riscrivendo abbiamo
$$
\begin{align}
    &P_e \cdot \log \Big(m - 1\Big) + 1 \geq H(X|Y) \quad &(1)\\
    &P_e \cdot \log \Big(m - 1\Big) \geq H(X|Y) - 1 \quad &(2)\\
    \\ 
    &\text{divido per il logaritmo}\\
    \\
    &P_e \geq \frac{H(X|Y) - 1}{\log \Big(|X| - 1\Big)} \quad &(3)
\end{align}
$$

<mark>Questo risultato ci permette di mostrare che lentropia  anche un **lower bound per la probabilit di errore**</mark>, oltre che per il valore atteso delle lunghezze del codice. Questo lower bound per  modulato sulla grandezza della sorgente.
