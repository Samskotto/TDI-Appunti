## Nozioni matematiche preliminari 
1. **Probabilit marginale**
    $$
        p(x) = \sum_{y \in Y} p(x,y)
    $$
2. **Probabilit condizionata**
   $$
   \begin{align}
       p(y|x) = \frac{p(x,y)}{p(x)} \quad p(x) > 0\\
       \\
       p(x|y) = \frac{p(x,y)}{p(y)} \quad p(x) > 0
    \end{align}
   $$
___
Abbiamo definito l'entropia su una singola variabile aleatoria, ma possiamo **estendere** questo concetto ad una **coppia di variabili aleatorie**.
Definiamo **l'entropia congiunta**come: 
$$
    H_D(X,Y) = \sum_{x \in X} \sum_{y \in Y} p(x,y) \cdot \log_D\frac{1}{p(x,y)}
$$
mentre **l'entropia condizionata**  definita come:
$$
\begin{align}
    H(Y|X) &= \sum_{x \in X} p(x) \cdot H(Y|X = x)\quad &(1)\\
    &= \sum_{x \in X}p(x) \cdot \sum_{y \in Y} p(y|x) \cdot \log \frac{1}{p(y|x)} \quad &(2)\\
    &= \sum_{x \in X}\sum_{y \in Y}p(x,y)\cdot\log \frac{1}{p(y|x)} \quad &(3)\\
\end{align}
$$
Nel passaggio $(2)$ si noti che $p(x) \cdot p(y|x) = p(x,y)$. 

*Semanticamente*: **l'entropia condizionata** indica la quantit di informazione media che serve per comunicare un messaggio estratto da $Y$ sapendo che ho gi mandato un messaggio estratto da $X$, mentre **l'entropia congiunta** indica la quantit di informazione che mi serve per descrivere due estrazioni da $X$ e $Y$. 
## Chain Rule
$$
    H(X,Y) = H(X) + H(Y|X)
$$
### <mark>Dimostrazione</mark>
Riscriviamo $H(X,Y)$:
$$
\begin{align}
    H(X,Y) &= \sum_{x \in X} \sum_{y \in Y} p(x,y) \cdot \log\frac{1}{p(x,y)} \quad &(1)\\
    &=-\sum_{x \in X} \sum_{y \in Y} p(x,y) \cdot \log p(x,y) \quad &(2)\\
    &=-\sum_{x \in X} \sum_{y \in Y} p(x,y) \cdot \log \Big[p(x)\cdot p(y|x)\Big] \quad &(3)\\
    &=-\sum_{x \in X} \sum_{y \in Y} p(x,y) \cdot \log p(x) - \sum_{x \in X} \sum_{y \in Y} p(x,y) \cdot \log p(y|x) \quad &(4)\\
    &=-\sum_{x \in X} p(x) \cdot \log p(x) - \sum_{x \in X} \sum_{y \in Y} p(x,y) \cdot \log p(y|x) \quad &(5)\\
    &=H(X) + H(Y|X) \quad &(6)\\
\end{align}
$$
Nel punto $(4)$ ho utilizzato le propriet dei logaritmi e si noti che <mark>$\sum_{y \in Y} p(x,y)$</mark>$\cdot \log p(x)$  la probabilit marginale $p(x)$.

Come corollario possiamo scrivere la **Chain rule** su uno spazio condizionato:
$$
    H(X,Y|Z) = H(X|Z) + H(Y|X,Z)
$$
___
## Informazione Mutua
Siano due variabili aleatorie $X$ e $Y$ con funzione di massa di probabilit congiunta $p(x,y)$ e funzioni di massa marginali $p(x)$ e $p(y)$. 
Linformazione mutua $I(X;Y)$  lentropia relativa tra la distribuzione congiunta e la distribuzione prodotto $p(x)\cdot p(y)$:
$$
    I(X;Y) = \sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log \frac{p(x,y)}{p(x)\cdot p(y)}
$$
La relazione che lega l'**informazione mutua** con l'**entropia**  data da:
$$
\begin{align}
    I(X;Y) = H(X) - H(X|Y) 
\end{align}
$$
<div style="background-color: #D1D7FF; padding: 10px; border-radius: 5px;text-align: center; font-size:16px;"><i>Semanticamente:</i> l'informazione mutua <mark>quantifica, in media, la dipendenza tra la distribuzione congiunta di $X$ e $Y$</mark> si discosta da quella che si otterrebbe se $X$ e $Y$ fossero <b>indipendenti</b>. Difatti se $X,Y$ sono <b>indipendenti</b> l'informazione mutua sarebbe pari $I(X;Y) = 0$ perch sapere $Y$ non ci da nessuna informazione sulla variabile $X$.
Un'altro modo di vederlo  che L'informazione mutua $I(X;Y)$  la riduzione dell'incertezza su $X$ dovuta alla conoscenza di $Y$.
</div>

### <mark> Dimostrazione </mark> 
Per definizione sappiamo che
$$
\begin{align}
    I(X;Y) &= \sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log \frac{p(x,y)}{p(x)\cdot p(y)}\quad &(1)\\
    &= \sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log \frac{p(y)\cdot p(x|y)}{p(x)\cdot p(y)} \quad &(2)\\
    &= \sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log \frac{p(x|y)}{p(x)} \quad &(3)\\
    &= -\sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log p(x|y)\cdot p(x) \quad &(4)\\
    &= -\sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log p(x) + \sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log p(x|y) \quad &(5)\\
    &= -\sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log p(x) - \Big(-\sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log p(x|y) \Big) \quad &(6)\\
    &= -\sum_{x \in X} p(x) \cdot \log p(x) - \Big(-\sum_{x \in X}\sum_{y \in Y} p(x,y) \cdot \log p(x|y) \Big) \quad &(7)\\
    &= H(X) - H(X|Y) \quad &(8)\\
\end{align}
$$

Applicando la **Chain rule** possiamo ottenere delle espressioni equivalenti


Anche **l'informazione mutua** possiamo vederla su uno spazio condizionato tale per cui:
$$
    I(X,Y|Z) = \sum_{x}\sum_{y}\sum_{z} p(x,y,z) \cdot \log \frac{p(x,y|z)}{p(x|z)\cdot p(y|z)}
$$
## Data Processing Inequality
Siano $X,Y,Z$ variabili aleatorie con condominio finito e che la loro distribuzione congiunta $p(x,y,z)$ soddisfa $p(x,z|y) = p(x|y) \cdot p(z|y)$ allora **l'informazione mutua** $I(X,Y) \geq I(X,Z)$

*Semanticamente*: questa disuguaglianza ci sta dicendo che se noi abbiamo questa catena $X \rightarrow Y \rightarrow Z$ se ottengo $Z$ grazie a $Y$ allora $Z$ non pu contenere pi informazione rispetto a $X$ (l'informazione inviata su un canale non pu essere maggiore del dato originale).
