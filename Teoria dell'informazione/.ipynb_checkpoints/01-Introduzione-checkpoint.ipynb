## Introduzione

**Claude Shannon** si è occupato di introdurre la *teoria dell'informazione* basata sulla probabilità. Negli anni 40 si occupa di studiare il problema della trasmissione di informazione su un canale fisico.

Shannon riesce a teorizzare questo problema utilizzando il teorema delle probabilità facendo però riferimento ad un **caso medio**, più tardi altri scienziati tra cui **Kolmogorov**, **Chaitin** e **Solomonoff**, avrebbero ripreso il suo lavoro cercando di risolvere i problemi aperti lasciati da Shannon rilasciando studi non più basati sulla teoria delle probabilità ma basati sulla *teoria della computazione*.

## Teoria dell'informazione secondo Shannon

Il problema che la teoria di Shannon affronta è quello di **trasmettere informazione attraverso un canale affetto da rumore**. Shannon quindi individua 5 attori principali:

Shannon decide di dividere ulteriormente il problema in due sottoproblemi:

1. **Source Coding** ($1\degree$Teorema di Shannon): trovare un codice per rappresentare il messaggio da inviare <mark>massimizzando la compressione</mark>, in modo da <mark>massimizzare l’informazione trasmessa ad ogni utilizzo del canale</mark>.

2. **Channel Coding** ($2\degree$Teorema di Shannon): trovare un codice per rappresentare il messaggio in modo da <mark>minimizzare gli errori di trasmissione dovuti al rumore che influisce sul canale</mark>.

### Rappresentazione Sorgente

Rappresentiamo la sorgente $\mathbb{X}$, come composta da un insieme finito di simboli, ad esempio:
\begin{align}
\mathbb{X} &= \{a, b, c,\dots \}\\
\text{oppure}\\
&\mathbb{X}= \{0, 1, 2, 3,\dots \}
\end{align}
Dall'insieme di simboli appartenenti alla sorgente $\mathbb{X}$, possiamo costruire un **messaggio** $x = (x_1,\dots,x_n)$, cioè una sequenza di $n$ simboli sorgente.
$$
    x \in \mathbb{X}^n \text{ dove } n = \text{lunghezza del messaggio}
$$

*Ad esempio* se consideriamo l'alfabeto italiano $\text{casa}\in \mathbb{X}^4$.

#### Definizione di Codifica
Sia $\mathbb{c}$ una funzione di codifica di base $D$ (esadecimale, ottale, binario, decimale$\dots$) che mappa i simboli della sorgente in parole di codice cioè in *sequenze di numeri* prese dall'$\{0, . . . , D−1\}^+$ tale per cui:
$$
    \mathbb{c}: \mathbb{X} \rightarrow \{0,\dots,D-1\}^+
$$
L'esponente $+$ all'insieme indica che consideriamo le sole sequenze $\geq 1$.
$$
\{0,\dots,D-1\}^+ = \bigcup_{i=1}^{\infty} \{0,\dots,D-1\}^i
$$
Stiamo unendo sostanzialmente tutte le sequenze di lunghezza $i$ possibili che possiamo costruire con base $D$, dato che l'unione parte da $i=1$ non stiamo considerando sequenze di lunghezza *vuota*.

#### Rappresentazione Canale

Rappresentiamo il canale con una **matrice stocastica** composta da:
1. In **input** abbiamo i simboli della sorgente
2. In **output** abbiamo la codifica del simbolo in input
3. i valori contenuti dalla matrice indicano la probabilità di ricevere il simbolo in presenza di rumore.
