## Algoritmo Sardinas - Patterson

Abbiamo come **input** l'insieme $S_1$ contenente le parole del codice $\mathbb{c}$.
1. Inizializziamo l'insieme $S_{i+1} = \emptyset$
2. Per ogni $x \in S_1$: se esiste $y$ tale che $xy \in S_i$, allora aggiungi $y \in S_{i+1}$.
3. Per ogni $x \in S_i$: se esiste $y$ tale che $xy \in S_1$, allora aggiungi $y \in S_{i+1}$.
4. Rimuoviamo eventuali duplicati da $S_{i+1}$, e verifichiamo i casi di terminazione.
5. Se nessun caso di terminazione si verifica, incremento $i$ e ripetiamo il ciclo dal punto 1.

I casi di **terminazione** sono:
- Se $S_{i+1} = \emptyset$ allora $\mathbb{c}$ è **univocamente decodificabile**
- Se $S_{i+1}$ è uguale ad almeno un insieme $S_j \quad t.c \quad j< i+1$ allora $\mathbb{c}$ è **univocamente decodificabile**
- Se $S_{i+1} \cap S_1 \neq \emptyset$ allora  $\mathbb{c}$ non è **univocamente decodificabile**
___
### Esempio 
- Sia $S_1 = \{A,E,C,ABB,CED,BBEC\}$ eseguiamo il primo passo per $i = 1$:
    1. $A \in S_1$ ed è testa della parola di codice $ABB$ quindi inseriamo in $S_{i+1} = S_2 = \{BB\}$
    2. $E \in S_1$ non è testa di nessuna parola
    3. $C \in S_1$ è testa della parola $CED$ quindi inseriamo in $S_2 = \{BB, ED\}$
    4. $ABB,CED,BBEC$ non sono testa di nessuna parola quindi **non inserisco** in $S_2$.
    Per $i = 1$ il punto numero 3 coincide con il punto numero 2 e le condizioni di terminazione non si verificano. 

- Per $i = 2$ costruiamo la sorgente $S_3$:
    1. $A \in S_1$ non è testa di nessuna parola in $S_2$
    2. $E \in S_1$ è testa di $ED \in S_2$ quindi inserisco in $S_3 = \{D\}$
    3. $C, ABB, CED, BBEC$ non sono testa di nessuna parola.
    4. Verifico ora la seconda condizione, cioè controllo sulle parole di $S_2 = \{BB,ED\}$ quali sono testa delle parole di $S_1$
        1. $BB \in S_2$ è testa di $BBEC \in S_1$ quindi aggiungo in $S_3 = \{D,EC\}$.
        2. $ED \in S_2$ non è testa di nessuna parola in $S_1$.
    A questo punto controllo i casi di terminazione

- Per $i = 3$ costruiamo la sorgente $S_4$:
    1. $A \in S_1$ non è testa, mentre $E \in S_1$ è testa di $EC \in S_3$ quindi aggiungo $S_4 = \{C\}$
    2. Termino, **il codice non è univocamente decodificabile** questo perchè ho trovato un insieme $S_4$ contenente una parola già presente in $S_1$.
