Siamo interessati a trovare anche un **upper bound** per il valore atteso delle lunghezze delle parole di un codice istantaneo. Abbiamo visto inoltre che il **codice di Shannon** gode di questa proprietà sulle lunghezze delle parole di codice:
$$
l_i = \lceil \log_{D}\frac{1}{p_i}\rceil
$$
## Teorema 
Per ogni sorgente $<\mathbb{X}, p>$ dove $\mathbb{X} = \{x_1,\dots,x_n\}$ e $p = (p_1,\dots,p_n)$ preso il codice istantaneo di **Shannon** $\mathbb{c}$ con lunghezze $l_i = l_c(x_i)$ tali che $l_i = \lceil \log_{D}\frac{1}{p_i}\rceil$ per $i = 1,\dots,n$ vale che 
$$
    E[l_c] < H_D(X) + 1
$$
Dato che lavoriamo in bits (quindi con $D = 2$) stiamo sostanzialmente dicendo che il valore atteso delle lunghezze del codice è minore dell'**entropia** più un **bit** di informazione.
### Dimostrazione
Sapendo che $E[l_c] = \sum_{i=1}^{n} p_i \cdot l_i$ noi sappiamo dato che stiamo utilizzando il codice istantaneo di **Shannon** che $l_i = \lceil \log \frac{1}{p_i} \rceil$ quindi riscriviamo:
$$
    E[l_c] = \sum_{i=1}^{n} p_i \cdot \lceil \log \frac{1}{p_i} \rceil
$$
Questa quantità è sicuramente minore di:
\begin{align}
    &\sum_{i=1}^{n} p_i \cdot \lceil \log \frac{1}{p_i} \rceil < \sum_{i=1}^{n} p_i \cdot \log_D \big(\frac{1}{p_i}\big) + 1 \quad &(1)\\
    &\sum_{i=1}^{n} p_i \cdot \lceil \log \frac{1}{p_i} \rceil < \sum_{i=1}^{n} p_i \cdot \log_D \big(\frac{1}{p_i}\big) + \sum_{i=1}^{n} p_i \quad &(2)\\
\end{align}
Ma nel punto $(2)$ abbiamo praticamente scritto la tesi che volevamo dimostrare cioè:
$$
 \sum_{i=1}^{n} p_i \cdot \lceil \log \frac{1}{p_i} \rceil < H_D(X) + 1
$$
___

Nel caso in cui utilizzassimo una variabile aleatoria $Y$ che cerca di stimare la distribuzione originale $X$ otteremmo questo risultato:

## Teorema
Dato un modello di sorgente $<\mathbb{X},p>$ se $\mathbb{c}$ è un codice di Shannon $\mathbb{c}: \mathbb{X}\rightarrow D^+$ con $l_c(x_i) = \lceil\log \frac{1}{q(x)}\rceil$ dove $q$ è una distribuzione arbitraria su $\mathbb{X}$ allora:
$$
    E[l_c] < H_D(X) + 1 + D(X || Y)
$$
Dove $X$ ha distribuzione $p$ e la variabile aleatoria $Y$ ha distribuzione $q$.

*Semanticamente* stiamo vedendo il caso in cui abbiamo una sorgente dove le lunghezze dipendono da una distribuzione $q$ e non da quella originale (quindi $q$ è una stima, per esempio la distribuzione $q$ può essere la migliore stima che possiamo fare della vera distribuzione sconosciuta)
### Dimostrazione
Riscriviamo $E[l_c] = \sum_{x \in \mathbb{X}} p(x) \cdot l_i$, dato che $l_i = \lceil\log \frac{1}{q(x)}\rceil$:
\begin{align}
    &\sum_{x \in \mathbb{X}} p(x) \cdot \lceil \log_D\frac{1}{q(x)}\rceil < \sum_{x \in \mathbb{X}} p(x) \cdot \big(\log\frac{1}{q(x)} + 1\big) \quad &(1)\\
    &\sum_{x \in \mathbb{X}} p(x) \cdot \lceil \log_D\frac{1}{q(x)}\rceil < \sum_{x \in \mathbb{X}} p(x) \cdot \big(\log\frac{1}{q(x)}\cdot \frac{p(x)}{p(x)} + 1\big) \quad &(2)\\
    &\sum_{x \in \mathbb{X}} p(x) \cdot \lceil \log_D\frac{1}{q(x)}\rceil < \sum_{x \in \mathbb{X}} p(x) \cdot \big(\log\frac{p(x)}{q(x)}+ \log \frac{1}{p(x)} + 1\big) \quad &(3)\\
    &\sum_{x \in \mathbb{X}} p(x) \cdot \lceil \log_D\frac{1}{q(x)}\rceil < \sum_{x \in \mathbb{X}} p(x)\cdot\log\frac{p(x)}{q(x)} +  \sum_{x \in \mathbb{X}} p(x)\log \frac{1}{p(x)} +  \sum_{x \in \mathbb{X}} p(x) \quad &(4)\\
\end{align}
Dal passaggio $(4)$ notiamo che:
\begin{align}
    \sum_{x \in \mathbb{X}} p(x)\cdot\log\frac{p(x)}{q(x)} &= D(X||Y)\\
    \sum_{x \in \mathbb{X}} p(x)\log \frac{1}{p(x)}  &= H(X)\\
    \sum_{x \in \mathbb{X}} p(x) &= 1
\end{align}    
quindi abbiamo dimostrato che $E[l_c] < D(X||Y) + H(X) + 1$.
