Abbiamo visto come possiamo rilevare l'occorenza di un singolo errore. Ma se vogliamo rilevare la presenza di un **numero maggiore di errori in un canale rumoroso**, cosa possiamo fare nella **progettazione dei codici** di rilevamento degli errori?

Supponiamo di voler spedire $3$-bit di informazione $(x_1,x_2,x_3)$ aggiungendo al messaggio $3$-bit di **parità** ulteriori di controllo tale per cui il messaggio trasmesso abbia questa forma $(x_1,x_2,x_3 \mid x_4,x_5,x_6)$. Siano definiti i bit di controllo in questo modo:
$$
\begin{cases}
    x_4 &= x_1 + x_2 \\
    x_5 &= x_1 + x_3 \\
    x_6 &= x_2 + x_3
\end{cases}
$$
Se il ricevente ottiene $(y_1,y_2,y_3 \mid y_4, y_5, y_6)$ e se viene violata anche una sola di queste condizioni:
$$
\begin{cases}
    y_4 &= y_1 + y_2 \\
    y_5 &= y_1 + y_3 \\
    y_6 &= y_2 + y_3
\end{cases}
$$
Viene rivelata l'occorenza di un errore, inoltre **possiamo ricavare la posizione $i$-esima** in cui **l'errore si è presentato**:
$$
\begin{cases}
    y_4 \neq y_1 + y_2, y_5 \neq y_1 + y_3 \quad &i = 1 \\
    y_4 \neq y_1 + y_2, y_6 \neq y_2 + y_3 \quad &i = 2 \\
    y_5 \neq y_1 + y_3, y_6 \neq y_2 + y_3 \quad &i = 3 \\
    y_4 \neq y_1 + y_2 \quad &i = 4 \\
    y_5 \neq y_1 + y_3 \quad &i = 5 \\
    y_6 \neq y_2 + y_3 \quad &i = 6 \\
\end{cases}
$$
Ma possiamo rilevare anche quando si **presentano doppi errori**.
