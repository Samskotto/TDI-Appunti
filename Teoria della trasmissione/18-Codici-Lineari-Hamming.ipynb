Definiamo un codice di correzione degli errori $\mathbb{c}$ di tipo $(n, k)$ che mappa $k$-bit (bit d'informazione della parola $s$) in $n$-bit della parola $x$. Diciamo che $\mathbb{c}$ è un codice lineare se:
$$
\exists G_{k\cdot n} \quad \text{e} \quad H_{(n-k)\cdot n} \quad \text{tali che}\quad  s \rightarrow x = (x_1,\dots,x_n) = (s_1,\dots,s_k) \cdot G
$$
E il controllo lato ricevente sulla correttezza del dato, sulla parola ricevuta può essere fatto moltiplicando: 
$$
    H \cdot x^t = (0,0, \dots, 0)
$$
Sostanzialmente $s$ è il messaggio da spedire sul canale, gli aggiungiamo **i bit di controllo** e otteniamo $x$ che verrà spedito sul canale. Il controllo lato ricevente sarà la matrice di parità moltiplicato per il vettore trasposto $x$, se è pari a zero allora il dato non è corrotto.

Un esempio di codice di questo tipo, è il **codice di Hamming** che è di tipo $(7,4)$ quindi usa parole di codice da $7$-bit ($3$-bit di controllo e $4$-bit di informazione). Lui costruisce questo codice per risolvere i problemi che riscontrava con l'utilizzo di schede perforate, perchè la macchina che leggeva l'assenza o la presenza dei fori nella scheda commetteva degli errori.
$$
\begin{cases}
    p_1 = s_1 + s_3 + s_4 \\
    p_2 = s_1 + s_2 + s_3 \\
    p_3 = s_2 + s_3 + s_4 
\end{cases}
$$
Possiamo ricavarci il sistema di equazioni attraverso **la matrice di parità** $H$ di dimensioni $(n-k)\times n = 3 \times 7$.
$$
H = \left[\begin{matrix}1 &0 &0 \mid &1 &0 &1 &1\\0 &1 &0 \mid &1 &1 &1 &0\\0 &0 &1 \mid &0 &1 &1 &1\\\end{matrix}\right]
$$
Da questa matrice possiamo ricavarci $G$ cioè la matrice generatrice di dimensioni $k \times n = 4 \times 7$ che è la sottomatrice trasposta
$$
G = \left[\begin{matrix}1 &1 &0 \mid &1 &0 &0 &0\\0 &1 &1 \mid &0 &1 &0 &0\\1 &1 &1 \mid &0 &0 &1 &0\\1 &0 &1 \mid &0 &0 &0 &1  \end{matrix}\right]
$$
Vediamo ora un esempio, sia questo vettore $(0,1,1 \mid 1,1,0,0)$ questa è la parola di codice che ho ricevuto, controlliamo la correttezza di questa parola di codice:
$$
  H \cdot \left(\begin{matrix}0\\1\\1\\1\\1\\0\\0\end{matrix}\right) =  \left(\begin{matrix}1\\1\\0\end{matrix}\right) \neq \left(\begin{matrix}0\\0\\0\end{matrix}\right)
$$
Il vettore che otteniamo ci da anche informazione su **dove** si è verificato l'errore, ci da la colonna corrispettiva: $s_1$ è il bit nel quale abbiamo commesso l'errore (ricordarsi che **Hamming** è in grado di correggere un singolo errore).

Contiamo il numero di $1$-all'interno degli insiemi, quelli che contengo un numero dispari di $1$-indicano gli errori, e l'intersezione (vediamo che in $s_1$ è presente un $1$) tra i due insiemi ci indica che l'errore è avvenuto in $s_1$.

Ad esempio con un vettore $(111\mid 0100)$ possiamo utilizzare **Mc-Lease** e vedere che:
$$
 p_1 = \{1, 0, 0, 0\} \quad p_2 = \{1, 0, 0, 1\} \quad p_3 =\{1, 1, 0, 0\}
$$
L'errore è quindi presente sul bit di controllo $p_1$.

Il **codice di Hamming** $(7,4)$ ci permette di correggere al più $1$ errore e di rilevare al più $2$-errori: perchè è in grado di **rilevarne due**?
$$
    H \cdot (y + e_1 + e_2)^{T} = H\cdot y^T + H\cdot e_1^T + H \cdot e_2^T =  H\cdot e_1^T + H \cdot e_2^T
$$
La **matrice di parità per costruzione**, se prendiamo **due colonne scelte arbitrariamente non avrò mai che la loro somma sia pari a zero**.
___
La **distanza di Hamming** è molto importante quando utilizziamo bit, questa distanza è molto comoda perchè se abbiamo un codice di tipo $(n,k)$ se prendiamo tra tutte le coppie di **code-word** quella che ha una distanza di Hamming minima pari a $d$, allora diciamo che il codice $\mathbb{c}$ costruito corregge $t$ errori:
$$
 t = \lfloor \frac{d-1}{2} \rfloor
$$
E rileva $d-1$ errori.
