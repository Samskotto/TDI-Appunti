Abbiamo visto i codici di **Hamming, Cilici, e Lineari** e abbiamo visto che possiamo utilizzarli per andare ad **identificare** gli errori attraverso la **matrice di parità**, e grazie a questo posso andare a correggere in presenza di **un singolo errore**. <mark>Utilizzo questo tipo di codici, quando so che il numero di errori è tenuto sotto controllo dalla probabilità di errore che il canale mi genera</mark>.

Ad esempio se so che in media ogni $100$ schede perforate una viene sbagliata e al più su quella scheda capita un singolo errore potrei usare **il codice di Hamming**.
Ora vediamo un'altra tipologia di codici: fino adesso abbiamo visto un canale con una probabilità di errore, e dato questo abbiamo scelto il codice adatto per controllare questo errore. I **codici BCH** permettono di correggere $t$-errori, in base a questo parametro $t$ utilizza e sceglie un codice adatto.

Per spiegare i **codici BCH** abbiamo bisogno di introdurre che cos'è un **campo di Galois**: i nostri computer per effettuare una divisione utilizzando questa operazione
$$
\frac{b}{a} = b \cdot a^{-1}
$$
Se noi possediamo un' anello $\mathbb{Z}_n$ stiamo lavorando su un'insieme finito di elementi, ma non è detto che tutti abbiamo l'inverso moltiplicativo che ci serve per eseguire le divisioni. Per questo motivo ci serve lavorare su un'campo $GF(p^m)$ che contiene un numero di elementi uguale a $p^m$ e ognuno di questi è un polinomio che ha grado massimo $m-1$ a coefficenti in $\mathbb{Z}_p$.
