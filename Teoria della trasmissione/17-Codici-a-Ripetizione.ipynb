## Codici a ripetizione tripla
Supponiamo di voler memorizzare un $1$-bit $s$. Dato che vogliamo essere in grado di recuperare il dato corrotto aggiungiamo quante più copie possiamo di $s$. Nel caso in cui ne memorizziamo $3$ siamo in presenza dello schema di un **codice a ripetizone tripla**.
$$
\begin{cases}
    x_2 = x_3 = 0 \quad \text{se}\quad x_1 = 0\\
    x_2 = x_3 = 1 \quad \text{se}\quad x_1 = 1\\
\end{cases}
$$
Quindi i possibili valori che può assumere sono $(0,0,0)$ o $(1,1,1)$. Per verificare che il **dato non sia corrotto** ci basta considerare questa condizione:
$$
\begin{cases}
    \text{Dato ok} \quad x_1+x_2 = 0 \text{ e } x_1+x_3 = 0\\
    \text{Dato corrotto, altrimenti}
\end{cases}
$$
Ora calcoliamo la probabilità $p(w(000)) = p(w(111)) = \frac{1}{2}$ (abbiamo  equiprobabilità di spedire $0,1$). Ora calcoliamo la probabilità di spedire $(000)$ e di aver ricevuto $(001)$:
$$
\begin{align}
    p(w(000)) \cap p(r(001)) &= p(w(000)) \cdot p(r(001)| w(000)) \quad &(1)\\
    &= \frac{1}{2} \cdot (1-p)\cdot(1-p)\cdot p \quad &(2)\\
    &= \frac{1}{2} \cdot (1-p)^2\cdot p \quad &(3)\\
\end{align}
$$
Nel punto $(2)$ vado a calcolare la proabilità di aver spedito $(0,0,0)$ ma di aver ricevuto $(0,0,1)$ come le probabilità che in prima/seconda posizione non ci sia stato un errore (quindi $1-p$) e che nella terza invece si presenti l'errore (quindi $p$).

Analogalmente possiamo ricavarci questa probabilità $p(w(111)) \cap p(r(001)) = \frac{1}{2}\cdot p^2(1-p)$.

Si dice che **il codice a ripetizione tripla** è un codice **in grado di correggere un singolo errore**, cioè che può correggere tutti i **possibili errori su un singolo bit**. Se durante la lettura risultano errati almeno due dei tre bit, allora questo codice è destinato a prendere una decisione errata.

## Definizione Error-correcting code
Un codice $\mathbb{c}$ di tipo $(n,k)$ è detto di **correzzione degli errori** se mappa $k$-bit di informazione della parola $s$ in $n$-bit della parola $x$, con un **code rate** $R = \frac{k}{n}$. Diciamo che $\mathbb{c}$ è un codice in grado di correggere $t$-errori se $\mathbb{c}$ può correggere qualisasi insieme di $t$ o meno errori nel vettore ricevuto di lunghezza $n$. 
___
Noi abbiamo visto che per capire se il dato, **in un codice a ripetizione tripla**, era corrotto valutando le condizioni $x_1+x_2 = 0 \cap x_1+x_3 = 0$, ma possiamo effettuare questo check utilizzando le matrici, ad esempio se $(x_1,x_2,x_3) = (1,1,1)$: 
$$
\begin{align}
 H = \left[\begin{matrix}1 &1 &0 \\1 &0 &1\end{matrix}\right] \\
 G = (1, 1, 1)
\end{align}
$$
$H$ è la matrice di parità, mentre $G$ è il polinomio generatore, sostanzialmente se esistono questi due allora siamo in presenza di un codice lineare.
